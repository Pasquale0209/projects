```python
from sklearn.preprocessing import LabelEncoder

def encoder(data,features):
  data2=data.copy()
  for elem in features:
    if elem != 'Length' and elem!= 'Channel' and elem!="DS Channel":
      #This line creates an instance of the LabelEncoder class, which will be used to encode the categorical values into numerical values.
      label_encoder=LabelEncoder()
      data2[elem]=label_encoder.fit_transform(data2[elem].astype(str))
  data2[features]=data2[features].astype(float)
  return data2

encoded=encoder(data,features)

#This line creates a new DataFrame named burst and drops the 'Timestamp' column from the encoded dataset.
burst=encoded.drop(['Timestamp'],axis=1)
burst=burst.groupby('MAC Address')

burst=burst.first().reset_index()

display(burst)
```

#***CLUSTERING***

```python
from sklearn.metrics.cluster import v_measure_score
from sklearn.metrics import mean_squared_error

# Definition of the function to calculate the similarity between request and cluster features
def similarity(request_features, cluster_features, threshold):

    # Count the number of matching features between request and cluster features
    matching_features = sum(1 for req, clus in zip(request_features, cluster_features) if req == clus)
    # Check if the number of matching features is greater than or equal to the threshold
    return matching_features >= threshold

# Implementation of the online clustering algorithm
def online_clustering(burst, N):
    clusters = {}
    # Counter to assign cluster id
    cluster_id_counter = 0

    #Iterate over each row of the dataframe
    for index, row in burst.iterrows():
        request_features = row.values[1:9]  # Exclude MAC address and Label
        #print(request_features)
        assigned_cluster = None

        #Iterate over existing clusters
        for cluster_id, cluster_features in clusters.items():
            # Check similarity between request and cluster features
            if similarity(request_features, cluster_features, N):
                assigned_cluster = cluster_id
                break

        #If no existing cluster is found, create a new cluster
        if assigned_cluster is None:
            # Creation of the new cluster
            cluster_id_counter += 1
            assigned_cluster = cluster_id_counter
            clusters[assigned_cluster] = request_features

        # Assign the cluster ID to the current request in the dataframe
        burst.at[index, 'cluster_id'] = assigned_cluster

    return burst,cluster_id_counter

# Test different configurations of N and select the best one based on clustering metrics
def test_N_values(burst, N_values):
    best_N = None
    best_v_measure = 0
    best_error = float('inf')

    # Iterate over each value of N
    for N in N_values:
        clustered_burst,l= online_clustering(burst.copy(), N)
        # Compute true labels and predicted labels for evaluation
        true_labels = clustered_burst['Label']
        predicted_labels = clustered_burst['cluster_id']

        v_measure = v_measure_score(true_labels, predicted_labels)
        # Calculate the absolute error between the number of unique predicted clusters and true clusters
        error=abs(len((clustered_burst["cluster_id"].unique()))-len((clustered_burst["Label"].unique())))

        if v_measure > best_v_measure:
            best_v_measure = v_measure
            best_N = N
            best_error = error

    return best_N, best_v_measure, best_error,l

# Testing N values
m=7
N_values = range(m)  # Test con diversi valori di N
best_N, best_v_measure, best_error,l = test_N_values(burst, N_values)

print("Best N value:", best_N)
print("Best v-measure:", best_v_measure)
print("Related error:", best_error)
#print(l)
```

FUNCTION FOR THE HEATMAP

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a function to plot a heatmap of a cross-tabulation between two columns of a DataFrame
def plot_heatmap(df,column1,column2,colormap="Blues"):

  # Create a cross-tabulation of the two columns
  crosstab=pd.crosstab(df[column1],df[column2])

  # Set the theme and font scale for the plot
  sns.set_theme(style="whitegrid",font_scale=1)

  # Set the size of the plot
  plt.figure(figsize=(15,10))
  sns.heatmap(crosstab,annot=True,fmt="d",cmap=colormap)

  plt.title(f"Heatmap of {column1} vs {column2}")
  plt.xlabel(column2)
  plt.ylabel(column1)

  plt.show()
```

```python
# Import the new dataset from the folder 'challenge-dataset'
new_data_path = "/content/challenge-dataset/"
new_data_files = os.listdir(new_data_path)
new_data=pd.DataFrame()

for file in new_data_files:
  new_data=pd.concat([new_data,pd.read_csv(new_data_path+file,index_col=False)])
#display(new_data)

#convert the datetime
new_data['Timestamp'] = pd.to_datetime(new_data['Timestamp'],unit= 's')

#count all the NaN values per column
new_nan_count = new_data.isna().sum()
#print(new_nan_count)

#calculate the percentage of NaN values per column
new_total_raws = len(new_data)
new_percentage_nan = (new_nan_count/new_total_raws)*100
#print(new_percentage_nan)

#drop the columns containing a percentage of NaN values greater then 60%
new_column_to_drop = new_percentage_nan[new_percentage_nan>60].index
new_data.drop(columns=new_column_to_drop, inplace=True)

#This line calculates the number of unique elements in each column of the data dataset and stores the result in a variable called uniquen_elem.
new_uniquen_elem = new_data.nunique()
#print(new_uniquen_elem)

new_features=[]

#This line initiates a for loop that iterates over the indices of the uniquen_elem array
for i in range(0,len(new_uniquen_elem)-1):
  #This line checks if the number of unique elements in the column at index i (retrieved using uniquen_elem[i]) is less than 20.
  if new_uniquen_elem[i]<20:
    new_features.append(new_uniquen_elem.index[i])
#print(new_features)

from sklearn.preprocessing import LabelEncoder

new_encoded=encoder(new_data,new_features)

#This line creates a new DataFrame named burst by dropping the 'Timestamp' column from the encoded dataset.
new_burst=new_encoded.drop(['Timestamp'],axis=1)
new_burst=new_burst.groupby('MAC Address')

new_burst=new_burst.first().reset_index()

display(new_burst)
```

```python
from sklearn.metrics.cluster import v_measure_score
random.seed(2222)

# Function to split dataset into training and test sets
def split_train_test(data, test_size):

    devices = data['Label'].unique()
    # Select random devices for the test set
    test_devices = random.sample(list(devices), test_size)
    print(test_devices)
    # Create training data by excluding test devices
    train_data = data[~data['Label'].isin(test_devices)]
    test_data= data.copy()
    # Create test data containing only the selected test devices
    test_data = test_data[test_data['Label'].isin(test_devices)]
    return test_data

# Function to perform online clustering validation for different test set sizes
def validate_online_clustering(data, K_values):
    results = {'K': [], 'v_measure': [], 'error': []}

    for K in K_values:
        v_measure_scores = []
        errors = []

        for _ in range(5):  # Repeat 5 times for reliable results
            # Split data into training and test sets
            test_data = split_train_test(data, K)
            #best_N, best_v_measure, best_error,l = test_N_values(test_data, N_values)
            print(f"BEST N {best_N}")
            true_labels = test_data['Label']
            # Perform online clustering on the test data with the best N value
            new_clustered_data,l=online_clustering(test_data, best_N)
            predicted_labels = new_clustered_data['cluster_id']
            print('Clusters are:')
            print(predicted_labels.unique())
            #print('******')
            #print(len(predicted_labels))

            v_measure = v_measure_score(true_labels, predicted_labels)
            v_measure_scores.append(v_measure)

            a = len(predicted_labels.unique())
            b = len(true_labels.unique())
            error = abs(a - b)
            errors.append(error)
            print('the v-measure is: ',v_measure)
            print('the error is: ',error)

        # Compute average v-measure score and error over 5 repetitions
        avg_v_measure = np.mean(v_measure_scores)
        avg_error = np.mean(errors)

        results['K'].append(K)
        results['v_measure'].append(avg_v_measure)
        results['error'].append(avg_error)

    return results

# Define K values for test set sizes
K_values = [2, 3, 4, 5]

# Perform online clustering validation for different test set sizes
validation_results = validate_online_clustering(new_burst, K_values)
#print(validation_results)

# Plot average performance measures vs K
plt.plot(validation_results['K'], validation_results['v_measure'], label='Average V-Measure')
plt.xlabel('Test Set Size (K)')
plt.ylabel('V-Measure')
plt.title('Average V-Measure vs Test Set Size (K)')
plt.legend()
plt.show()
```

```python
plt.xlabel('Test Set Size (K)')
plt.ylabel('Error')
plt.title('Average Error vs Test Set Size (K)')
plt.plot(validation_results['K'], validation_results['error'], label='Average Error')
plt.legend()
plt.show()
```

The different test sets are composed by k randomly selected devices. As can be seen in the figure by increasing the test size, the number of clusters increases,the average error increases, this leads to a decrease in completeness while the homogeneity remains almost the same and so the v-measure, that is the harmonic mean between homogeneity and completeness, should decrease. This is due to the fact that when we increase the test size, we are increasing the number of devices that should be recognized through the clustering algorithm and so consequently the precision in classifying them decreses.
